```python
#For loops
for i in range(10):
  print(i)
```

```python
for i in range(1,10,2): #using the step (start, end, step)
  print(i)
```

```python
for i in range(0, 10,2): #using the step (start, end, step)
  print(i)
```

```python
#for the reverse order
for i in range(10,1,-1):
  print(i)
```

```python
#for the reverse order
for i in range(10,1,-2): #using a negative stepsize (even numbers)
  print(i)
```

```python
#for the reverse order
for i in range(9,0,-2): #using a negative stepsize ( odd numbers)
  print(i)
```

```python
# For loops with strings
name ="vinay"
for i in name:
  print(i)
```

While Loops

Continues as long as the condition is true

```python
#while lopps with range
i = 0
while i < 5:
  print(i)
  i = i+1
```

Loop control statements

```python
name = "vinay"
while "a" in name:
  print(name)
  break #helps to break the loop
```

```python
for i in range(10):
  if i==6:
    break
  print(i)
```

```python
#continue statement: skips the current iteration and moves to the nextone
for i in range(10):
  if i==6:
    continue #here it skips printing 6
  print(i)
```

```python
#PAss statements, which just skips the current line and passes on to the next one
for i in range(7):
  if i==6:
    pass #here it passes to the next line and prints 6
  print(i)
```

```python
#Nested Loops
for i in range(3):
  for j in range(4):
    print("{} and {}".format(i,j)) # can also be written using fstrings(see below)
```

```python
#Nested Loops
for i in range(3):
  for j in range(4):
    print(f"{i} and {j}") # using f string
```

```python
#calculate the sum of n numbers using while and for loop

n = int(input("enter the number "))

sum = 0
while(n!=0):
  sum = sum + n
  n = n-1
print("the total sum is ", sum)
```

```python
#using for loop
n = int(input("enter the number "))
sum = 0

for i in range(n,0,-1):
  sum = sum + i
print("the total sum is ", sum)
```

```python
#prime numbers between 1 and 100
num = 100
for num in range(1,num+1 ):
  if num > 1:
    for i in range(2, num):
      if num%i == 0:
        break
    else:
      print(num)
```
