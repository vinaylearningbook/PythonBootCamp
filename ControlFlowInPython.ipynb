A Simple Calculator

```python
x = float(input("Enter the first number"))
y = float(input("Enter the second number"))
opr = input("Enter the operation ('+', '-', '*', '/')")

if opr=="+":
  print("sum of {} and {} is {}".format(int(x),int(y), int(x+y)))
elif opr=="-":
  print("difference of {} and {} is {}".format(int(x),int(y), int(x-y)))
elif opr=="*":
  print("product of {} and {} is {}".format(int(x),int(y), int(x*y)))
elif opr=="/":
  if y!=0:
    print("division of {} and {} is {}".format(int(x),int(y), int(x/y)))
  else:
    print("division by zero is not possible")
else:
  print("Invalid operation")
```

Determine the ticket price based on age

```python
x = int(input("Enter your age "))
y = str(input("are you a student (yes/no)? "))

def isStudent(z):
    if z.lower() == "yes":
        return True
    else:
        return False

if x > 0 and x < 5:
    print("your ticket is free")
elif x < 13:
    print("you're a minor hence your ticket price will be 5$")
elif x < 18:
    if isStudent(y):
        print("you're still a student so your ticket price is 10$")
    else:
        print("you're ticket price would be 15$ as you're a teenager")
elif x > 18:
    print("you're an adult hence the ticket price is 25$")
else:
    print("Invalid age")
```
