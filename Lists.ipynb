List Comprehensions

```python
#old way or begineer way
lst = []
for i in range(10):
  lst.append(i**2)
print(lst)
```

```python
#all the above can be dine in a single line inside the list
lst = [i**2 for i in range(10)]
print(lst)
```

```python
##### List comprehension syntax
#[expression for item in iterable]
```

#List comprehension syntax
### Basic Syntax [expression for item in iterable]
### Syntax with condition: [expression for item in iterable if condition]
### Nested list comprehension [expression for item1 in iterable1 item2 in iterable2]

```python
#all the above can be dine in a single line inside the list
lst = [i**2 for i in range(10)]
print(lst)
```

```python
#all the above can be dine in a single line inside the list
lst = [i**2 for i in range(10) if i%2==0] # only even numbers
print(lst)
```

```python
#Nested list comprehension

lst1 = [1,2,3]
lst2 = [4,5,6]
lst3 = [i*j for i in lst1 for j in lst2]
print(lst3)
```

```python
#all the above can be dine in a single line inside the list
lst = [i**2 for i in range(10) if i%2==0 else i**1] # only even numbers
print(lst)
```
