Lists and List Comprehensions

Lists are ordered and mutable collection of items

They can contain items of different data types

```python
lst = []
print(type(lst))
```

```python
lst = ["apple", 1, True, 3.24, "banana"] #can hold any type of elements
print(lst)
```

```python
#going through the listusing for loop
for i in range(len(lst)):
  print(lst[i])
```

```python
#accessing via index value
print(lst[4])
print(lst[0])
print(lst[3])
print(lst[-1]) #reverse indexing -1 representing the last entry
print(lst[-2])
```

```python
#slicing
print(lst[1:]) #skips the first element
print(lst[1:3])
print(lst[-4:-1]) # must be used in the reverse order
```
